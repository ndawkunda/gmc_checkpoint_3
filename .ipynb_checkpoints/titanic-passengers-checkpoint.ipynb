```python
df['Age'].fillna(df['Age'].mean(), inplace=True)
df['Embarked'].fillna('S', inplace=True)
df.isnull().sum().sum()
```

```python
def plot_correlation_map(df):
    corr = df.corr()
    s , ax = plt.subplots(figsize = (12 , 10))
    cmap = sns.diverging_palette(220 , 10 , as_cmap = True)
    s = sns.heatmap(
        corr, 
        cmap = cmap,
        square=True, 
        cbar_kws={ 'shrink' : .9 }, 
        ax=ax, 
        annot = True, 
        annot_kws = { 'fontsize' : 12 }
        )
plot_correlation_map(df)
```

This function takes a dataframe as a parameter and then generates a heatmap that shows the correlation between the different columns of the dataframe, the study of this graph will allow us to identify the relationships that exist between these columns and to select the features that have the most impact on the target. We can observe in our case certain correlations which could be useful to us:
- Pclass and Fare: correlation coefficient = -0.55
- Survived and Sex: correlation coefficient = -0.54
- Survived and Pclass: correlation coefficient = -034
- SibSp and Parch: correlation coefficient = 0.41
